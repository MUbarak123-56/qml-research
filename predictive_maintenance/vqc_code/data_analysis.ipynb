```python
import pandas as pd
import numpy as np
import seaborn as sns; sns.set_style("darkgrid")
import matplotlib.pyplot as plt
%matplotlib inline
import os
import ast
```

```python
res_list = os.listdir("../vqc_results/train_process")
```

```python
df = pd.DataFrame()
for c in res_list:
    if ".csv" in c:
        n_df = pd.read_csv("../vqc_results/train_process/" + c)
        n_df["reps"] = c.split("_")[-1][0]
        df = pd.concat([df, n_df], axis = 0).reset_index(drop =True)
#df
```

```python
tot_df = df
```

```python
#tot_df["reps"] = tot_df["reps"].fillna(2.0)
tot_df
```

```python
tot_df.sort_values("f1_test", ascending=False)
```

```python
tot_df = tot_df[tot_df["reps"] == 2.0]
```

```python
n = []
n.append(ast.literal_eval(tot_df.loc[0, "objective_vals"]))
#n
ansatz_list = list(tot_df["Ansatz"].unique())
optim_list = list(tot_df["optimizer"].unique())
ansatz_list, optim_list
```

### Loss Analysis

```python
def loss_eda(feat_map, col="blue"):
    data = tot_df[tot_df["feature_map_type"] == feat_map]
    fig, axes = plt.subplots(nrows=len(ansatz_list), ncols=len(optim_list), figsize = (28, 16))
    for i in range(len(ansatz_list)):
        for j in range(len(optim_list)):
            use_data = data[(data["Ansatz"] == ansatz_list[i]) & (data["optimizer"] == optim_list[j])].reset_index(drop=True)
            loss_val = ast.literal_eval(use_data.loc[0,"objective_vals"])
            #print(loss_val)
            sns.lineplot(x = range(len(loss_val)), y = loss_val, ax = axes[i,j], color=col)
            ax = axes[i,j]
            ax.set_ylabel("loss", fontsize=10)
            ax.set_xlabel("iteration", fontsize=10)
            ax.set_title("Loss plot using %s and %s"%(ansatz_list[i], optim_list[j]))
```

```python
loss_eda("Pauli","red")
```

## Univariate Analysis

```python
def box_relation(cat, num):
    plt.figure(figsize=(15,8))
    sns.boxplot(x=cat, y = num, data = tot_df)
    plt.xlabel(cat)
    plt.ylabel(num)
    plt.title("Distribution of %s across %s"%(num, cat))
    plt.show();  
```

```python
box_relation("Ansatz", "Training time")
```

```python
box_relation("Ansatz", "f1_score")
```

```python
box_relation("optimizer", "Training time")
```

```python
box_relation("optimizer", "f1_score")
```

## Multivariate Analysis

```python
def conf_mat_create(feat_map, var1, var2, num):
    use_df = tot_df[tot_df["feature_map_type"]==feat_map]
    use_df = use_df[[var1, var2, num]]
    pivot_df = use_df.pivot_table(index = [var1], columns = var2, values=num)
    plt.figure(figsize=(15,10))
    sns.heatmap(pivot_df, annot=True, fmt=".2f", linewidths=.5, cmap="RdBu",)
    plt.title("%s across different %s and %s combos for %s feature map"%(num, var1, var2, feat_map))
    #return pivot_df
```

```python
conf_mat_create("Pauli","optimizer", "Ansatz", "Training time")
```

```python
conf_mat_create("Pauli","optimizer", "Ansatz", "f1_score")
```
