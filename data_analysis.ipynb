```python
import pandas as pd
import os
```

```python
data_list = os.listdir("data_for_paper")
```

```python
performance_df = pd.DataFrame()
for i in range(len(data_list)):
    if "performance" in data_list[i]:
        df = pd.read_csv("data_for_paper/" + data_list[i])
        performance_df = pd.concat([performance_df, df], axis = 0).reset_index(drop=True)
```

```python
performance_df
```

```python
performance_df = performance_df[["f1_test", "f1_train", "prec_test", "prec_train", "recall_test", "recall_train", "model","elapsed", "data"]]
```

```python
performance_df.to_csv("collected_data/performance.csv", index=False)
```
