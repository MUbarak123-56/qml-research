### Univariate analysis

```python
def cat_eda(var):
    plt.figure(figsize=(15,10))
    sns.countplot(x = var, data = train, hue  = "target")
    plt.show();
```

```python
cat_eda("type")
```

```python
def num_eda(var):
    plt.figure(figsize=(15,10))
    sns.kdeplot(x = var, data = train, hue  = "target")
    #plt.ylim(0, 500)
    plt.show(); 
```

### Multivariate Analysis

```python
plt.figure(figsize=(40,20))
sns.pairplot(train_use, hue="target")
plt.show();
```

```python
cat_col = []
num_col = []
for col in list(train.columns):
    if is_categorical_dtype(train[col]):
        cat_col.append(col)
    elif is_numeric_dtype(train[col]):
        num_col.append(col)
```

```python
num_col, cat_col
```

```python
num_col.remove("target")
num_col.remove("udi")
```

```python
num_col, cat_col
```

```python
def multiple_cat_num_eda(num, cat):
    fig, axes = plt.subplots(nrows=len(num), ncols=1, figsize = (20, 40))
    for i in range(len(num)):
        for j in range(len(cat)):
          #if var_list[i] != var_list[j]:
            sns.violinplot(x = cat[j], y = num[i], data = train, hue = "target", ax = axes[i])
            ax = axes[i]
            ax.set_ylabel(num[i], fontsize=14)
            ax.set_xlabel(cat[j], fontsize=14)
```

```python
multiple_cat_num_eda(num_col, cat_col)
```
