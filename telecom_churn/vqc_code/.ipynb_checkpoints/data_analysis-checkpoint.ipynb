```python
import pandas as pd
import numpy as np
import seaborn as sns; sns.set_style("darkgrid")
import matplotlib.pyplot as plt
%matplotlib inline
import os
import ast
```

```python
custom = os.listdir("../vqc_results/custom")
zz = os.listdir("../vqc_results/zz")
pauli = os.listdir("../vqc_results/pauli")
```

```python
df = pd.DataFrame()
for c in custom:
    if ".csv" in c:
        n_df = pd.read_csv("../vqc_results/custom/" + c)
        df = pd.concat([df, n_df], axis = 0).reset_index(drop =True)
#df
```

```python
df2 = pd.DataFrame()
for z in zz:
    if ".csv" in z:
        n_df2 = pd.read_csv("../vqc_results/zz/" + z)
        df2 = pd.concat([df2, n_df2], axis = 0).reset_index(drop = True)
#df2
```

```python
df3 = pd.DataFrame()
for p in pauli:
    if ".csv" in p:
        n_df3 = pd.read_csv("../vqc_results/pauli/" + p)
        df3 = pd.concat([df3, n_df3], axis = 0).reset_index(drop = True)
#df3
```

```python
tot_df = pd.concat([df, df2, df3], axis = 0).reset_index(drop=True)
tot_df.info()
```

```python
n = []
n.append(ast.literal_eval(tot_df.loc[0, "objective_vals"]))
#n
ansatz_list = list(tot_df["Ansatz"].unique())
optim_list = list(tot_df["optimizer"].unique())
ansatz_list, optim_list
```

### Loss Analysis

```python
def loss_eda(feat_map, col="blue"):
    data = tot_df[tot_df["feature_map_type"] == feat_map]
    fig, axes = plt.subplots(nrows=len(ansatz_list), ncols=len(optim_list), figsize = (28, 16))
    for i in range(len(ansatz_list)):
        for j in range(len(optim_list)):
            use_data = data[(data["Ansatz"] == ansatz_list[i]) & (data["optimizer"] == optim_list[j])].reset_index(drop=True)
            loss_val = ast.literal_eval(use_data.loc[0,"objective_vals"])
            #print(loss_val)
            sns.lineplot(x = range(len(loss_val)), y = loss_val, ax = axes[i,j], color=col)
            ax = axes[i,j]
            ax.set_ylabel("loss", fontsize=10)
            ax.set_xlabel("iteration", fontsize=10)
            ax.set_title("%s loss plot using %s and %s"%(feat_map, ansatz_list[i], optim_list[j]))
```

```python
loss_eda("Custom","red")
```

```python
loss_eda("ZZ", "green")
```

```python
loss_eda("Pauli")
```

## Univariate Analysis

```python
def box_relation(cat, num):
    plt.figure(figsize=(15,8))
    sns.boxplot(x=cat, y = num, data = tot_df)
    plt.xlabel(cat)
    plt.ylabel(num)
    plt.title("Distribution of %s across %s"%(num, cat))
    plt.show();  
```

```python
box_relation("feature_map_type", "Training time")
```

## Multivariate Analysis

```python
def conf_mat_create(feat_map, var1, var2, num):
    use_df = tot_df[tot_df["feature_map_type"]==feat_map]
    use_df = use_df[[var1, var2, num]]
    pivot_df = use_df.pivot_table(index = [var1], columns = var2, values=num)
    plt.figure(figsize=(15,10))
    sns.heatmap(pivot_df, annot=True, fmt=".2f", linewidths=.5, cmap="RdBu")
    plt.title("%s across different %s and %s combos for %s feature map"%(num, var1, var2, feat_map))
    #return pivot_df
```

```python
conf_mat_create("Custom","optimizer", "Ansatz", "Training time")
```

```python
conf_mat_create("Custom","optimizer", "Ansatz", "f1_score")
```

```python
conf_mat_create("ZZ","optimizer", "Ansatz", "Training time")
```

```python
conf_mat_create("ZZ","optimizer", "Ansatz", "f1_score")
```

```python
conf_mat_create("Pauli","optimizer", "Ansatz", "Training time")
```

```python
conf_mat_create("Pauli","optimizer", "Ansatz", "f1_score")
```

```python
plt.figure(figsize=(13,7))
sns.scatterplot(x="Training time", y="f1_score", data=tot_df, hue = "feature_map_type")
plt.xlabel("Training time")
plt.ylabel("F1 score")
plt.title("F1 score vs Training time")
plt.text(x = 10000, y=0.86,s = "Best Algo", fontdict={"fontweight":"bold"})
plt.legend(loc="best")
plt.show()
```

### Data Extraction

```python
def extract(feature_map):
    new_df = tot_df[tot_df["feature_map_type"]==feature_map]
    return new_df[["feature_map_type", "optimizer", "Ansatz", "f1_score", "Training time"]].sort_values("f1_score", ascending=False)
```
