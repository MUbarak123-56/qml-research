```python
import pandas as pd
import numpy as np
import seaborn as sns; sns.set_style("darkgrid")
import matplotlib.pyplot as plt
%matplotlib inline
import os
import ast
```

```python
res_list = os.listdir("../vqc_results/train_process")
```

```python
df = pd.DataFrame()
for c in res_list:
    if ".csv" in c:
        n_df = pd.read_csv("../vqc_results/train_process/" + c)
        n_df["reps"] = c.split("_")[-1][0]
        df = pd.concat([df, n_df], axis = 0).reset_index(drop =True)
#df
```

```python
tot_df = df
```

```python
best_df = tot_df.sort_values("f1_test", ascending=False).reset_index(drop=True)
```

```python
best_df["reps"] = best_df["reps"].astype(int)
```

```python
#loss_val=best_df.loc[0,"objective_vals"]
loss_val = ast.literal_eval(best_df.loc[0,"objective_vals"])
```

```python
plt.figure(figsize=(16,8))
sns.lineplot(x = range(len(loss_val)), y = loss_val, color="blue")
plt.ylabel("Loss", fontsize=10)
plt.xlabel("Iteration", fontsize=10)
plt.title("Cost function for VQC")
plt.savefig("../../images/vqc_results.png")
```
