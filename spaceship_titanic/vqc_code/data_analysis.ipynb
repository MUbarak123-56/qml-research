```python
res_list = os.listdir("../vqc_results/train_process")
```

```python
df = pd.DataFrame()
for c in res_list:
    if ".csv" in c:
        n_df = pd.read_csv("../vqc_results/train_process/" + c)
        n_df["reps"] = c.split("_")[-1][0]
        df = pd.concat([df, n_df], axis = 0).reset_index(drop =True)
#df
```

### Loss Analysis

```python
def loss_eda(feat_map, rep, col="blue"):
    new = tot_df[tot_df["reps"] == rep].reset_index(drop=True)
    data = new[new["feature_map_type"] == feat_map].reset_index(drop=True)
    fig, axes = plt.subplots(nrows=len(ansatz_list), ncols=len(optim_list), figsize = (24, 16))
    for i in range(len(ansatz_list)):
        for j in range(len(optim_list)):
            use_data = data[(data["Ansatz"] == ansatz_list[i]) & (data["optimizer"] == optim_list[j])].reset_index(drop=True)
            loss_val = ast.literal_eval(use_data.loc[0,"objective_vals"])
            #print(loss_val)
            sns.lineplot(x = range(len(loss_val)), y = loss_val, ax = axes[i,j], color=col)
            ax = axes[i,j]
            ax.set_ylabel("loss", fontsize=10)
            ax.set_xlabel("iteration", fontsize=10)
            ax.set_title("Loss plot using %s and %s for %d reps"%(ansatz_list[i], optim_list[j], rep))
```

```python
loss_eda("Pauli", 1, "red")
```

```python
loss_eda("Pauli", 2, "blue")
```

```python
loss_eda("Pauli", 3, "purple")
```

## Univariate Analysis

```python
def box_relation(cat, num, rep):
    new = tot_df[tot_df["reps"] == rep].reset_index(drop=True)
    plt.figure(figsize=(15,8))
    sns.boxplot(x=cat, y = num, data = new)
    plt.xlabel(cat)
    plt.ylabel(num)
    plt.title("Distribution of %s across %s for %d reps"%(num, cat, rep))
    plt.show();  
```

```python
box_relation("Ansatz", "Training time", 1)
box_relation("Ansatz", "Training time", 2)
box_relation("Ansatz", "Training time", 3)
```

```python
box_relation("Ansatz", "f1_test", 1)
box_relation("Ansatz", "f1_test", 2)
box_relation("Ansatz", "f1_test", 3)
```

```python
#box_relation("optimizer", "Training time")
box_relation("optimizer", "Training time", 1)
box_relation("optimizer", "Training time", 2)
box_relation("optimizer", "Training time", 3)
```

```python
box_relation("optimizer", "f1_test", 1)
box_relation("optimizer", "f1_test", 2)
box_relation("optimizer", "f1_test", 3)
```

## Multivariate Analysis

```python
def conf_mat_create(feat_map, var1, var2, num, rep):
    use_df = tot_df[tot_df["feature_map_type"]==feat_map]
    use_df = use_df[use_df["reps"] == rep].reset_index(drop=True)
    use_df = use_df[[var1, var2, num]]
    pivot_df = use_df.pivot_table(index = [var1], columns = var2, values=num)
    plt.figure(figsize=(15,10))
    sns.heatmap(pivot_df, annot=True, fmt=".2f", linewidths=.5, cmap="RdBu")
    plt.title("%s across different %s and %s combos for %s feature map using %d reps"%(num, var1, var2, feat_map, rep))
    #return pivot_df
```

```python
conf_mat_create("Pauli","optimizer", "Ansatz", "Training time", 1)
conf_mat_create("Pauli","optimizer", "Ansatz", "Training time", 2)
conf_mat_create("Pauli","optimizer", "Ansatz", "Training time", 3)
```

```python
conf_mat_create("Pauli","optimizer", "Ansatz", "f1_test", 1)
conf_mat_create("Pauli","optimizer", "Ansatz", "f1_test", 2)
conf_mat_create("Pauli","optimizer", "Ansatz", "f1_test", 3)
```
